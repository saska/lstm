```python
rates = np.zeros((len(preds)-1, len(preds[0])))
for i in range(len(preds)-1):
    for j in range(len(preds[i])):
        if preds[i][j] < 0:
            rates[i][j] -= targets[i+1][j]
        else:
            rates[i][j] += targets[i+1][j]
```

```python
sumrates = rates.sum(axis=1) + 1
```

```python
assert(np.array(abs(rates)) == np.array(abs(targets[1:])))
```

```python
np.testing.assert_array_equal(np.array(abs(rates)), abs(targets[1:]))
```
